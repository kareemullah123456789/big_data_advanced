```python
path = "dbfs:/databricks-datasets/online_retail/data-001/"

df = spark.read.option("header", "true").option("inferSchema", "true").csv(path)
display(df.limit(10))
```

```python
df_clean = df.na.drop(subset=["InvoiceNo", "StockCode", "Quantity", "UnitPrice"]) \
              .withColumn("TotalAmount", F.col("Quantity") * F.col("UnitPrice"))
display(df_clean.limit(5))

```

```python
from pyspark.sql import functions as F
import time

# Start timer
start = time.time()

# Perform aggregation
sales_country = (
    df_clean.groupBy("Country")
    .agg(F.sum("TotalAmount").alias("TotalSales"))
    .orderBy(F.desc("TotalSales"))
)

# Trigger computation & show results
display(sales_country)

# End timer
print("Time taken (s):", round(time.time() - start, 2))

```

```python
import time
from statistics import median

def time_query(fn, runs=3):
    times = []
    for _ in range(runs):
        start = time.time()
        fn()   # fn must trigger an action (e.g., display(), count(), collect())
        times.append(time.time() - start)
    return median(times), times

```
