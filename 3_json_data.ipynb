```python
import json  # Import the JSON module for parsing

# Define a JSON-formatted string
sample_json = """{
  "id": 143,
  "name": "Silicon Valley",
  "type": "Scripted",
  "language": "English",
  "genres": ["Comedy"],
  "network": {
    "id": 8,
    "name": "HBO",
    "country": {
      "name": "United States",
      "code": "US",
      "timezone": "America/New_York"
    }
  }
}"""

# Parse the JSON string into a Python dictionary
document = json.loads(sample_json)

# Print the parsed document (Python dictionary)
print(document)

# Verify the type of the parsed object
print(type(document))  # Output: <class 'dict'>
```

```python
from pyspark.sql import SparkSession

# Create a SparkSession (the entry point to Spark functionality)
spark = SparkSession.builder.getOrCreate()

# Use Spark's JSON reader to load data from a JSON file
# Note: The path "./data/shows/shows-silicon-valley.json" points to the input file
shows = spark.read.json("/content/drive/MyDrive/cde_data/shows/shows-silicon-valley.json")

# Count the number of records in the DataFrame
# Note: count() is an action that triggers actual computation
record_count = shows.count()

# Print the result
print(f"Number of records: {record_count}")
```

```python
shows.show(5,False)
```

```python
# Read multiple JSON files using a wildcard pattern
three_shows = spark.read.json(
    "/content/drive/MyDrive/cde_data/shows/shows-*.json",  # Wildcard (*) matches multiple files
    multiLine=True  # Required for JSON records spanning multiple lines
)

# Count the number of records and verify the assertion
record_count = three_shows.count()
print(f"Number of records: {record_count}")  # Output: 3
assert record_count == 3, "Expected exactly 3 records"
```
